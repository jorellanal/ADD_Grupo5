```python
import pandas as pd

# Ajuste de la ruta 'Files/' y el argumento header=4 para todos los archivos

# Archivo clave (Provincia y Tramo de Trabajadores)
df_prov_trtrab = pd.read_excel(
    'Files/PUB_PROV_TRTRAB.xlsb', 
    engine='pyxlsb',
    header=4
)

# Archivos Regionales
df_reg = pd.read_excel('Files/PUB_REG.xlsb', engine='pyxlsb', header=4)
df_reg_act = pd.read_excel('Files/PUB_REG_ACT.xlsb', engine='pyxlsb', header=4)
df_reg_rubr = pd.read_excel('Files/PUB_REG_RUBR.xlsb', engine='pyxlsb', header=4)
df_reg_subr = pd.read_excel('Files/PUB_REG_SUBR.xlsb', engine='pyxlsb', header=4)
df_reg_trtrab = pd.read_excel('Files/PUB_REG_TRTRAB.xlsb', engine='pyxlsb', header=4)

dataframes = {
    "df_prov_trtrab (PROVINCIA/TRAMO - CLAVE)": df_prov_trtrab,
    "df_reg (REGION)": df_reg,
    "df_reg_act (REGION/ACTIVIDAD)": df_reg_act,
    "df_reg_rubr (REGION/RUBRO)": df_reg_rubr,
    "df_reg_subr (REGION/SUBRUBRO)": df_reg_subr,
    "df_reg_trtrab (REGION/TRAMO)": df_reg_trtrab,
}

for name, df in dataframes.items():
    print("=" * 70)
    print(f"DataFrame: {name}")
    print("Columnas:", df.columns.tolist()[:5], "...") # Mostrar solo las 5 primeras columnas
    print("-" * 70)
    print(df.head(5).to_markdown(index=False, numalign="left", stralign="left"))



```
